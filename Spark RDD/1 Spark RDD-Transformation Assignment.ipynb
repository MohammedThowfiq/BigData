```
# ------------------------mapPartitionsWithIndex-------------------------
```

```
# Example 1

def process_partition_y_sum(index,partition):
  number_sum = 0
  for ele in partition:
    number_sum = number_sum + ele
  yield number_sum

rdd1 = sc.parallelize([1,2,3,1,3,4,1,2,5,1],3)
rdd_partitions = rdd1.mapPartitionsWithIndex(process_partition_y_sum)
rdd_partitions.collect()
```
